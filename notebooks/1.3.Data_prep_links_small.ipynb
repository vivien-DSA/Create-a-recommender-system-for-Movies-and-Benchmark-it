# 2. Preprocessing of the data 

This data contains the TMDB and IMDB IDs of a small subset of 9,000 movies of the Full Dataset.<br>
Columns contents:
- **movieId:** The id of the movie. Movie ids are consistent between `ratings_small.csv`, `movies_metadata.csv`, `credits.csv`, `keywords.csv`and `links_small.csv` (i.e., the same id refers to the same movie across these data files).
- **imdbId:** The identifier for the Internet Movie Database (IMDb) [with prefix 'tt', 'nm', 'co', 'ev', 'ch' or 'ni']. 
- **tmdbId:** Identifier for a movie at The Movie Database (TMDb).
